```python
import numpy as np
import matplotlib.pyplot as plt

def plot_stacked_bar(data, series_labels, category_labels=None,show_values=False, value_format="{}", y_label=None,colors=None, grid=True, reverse=False):
    ny = len(data[0])
    ind = list(range(ny))

    axes = []
    cum_size = np.zeros(ny)

    data = np.array(data)

    if reverse:
        data = np.flip(data, axis=1)
        category_labels = reversed(category_labels)

    for i, row_data in enumerate(data):
        color = colors[i] if colors is not None else None
        axes.append(plt.bar(ind, row_data, bottom=cum_size, 
                            label=series_labels[i], color=color))
        cum_size += row_data

    if category_labels:
        plt.xticks(ind, category_labels)

    if y_label:
        plt.ylabel(y_label)

    plt.legend()

    if grid:
        plt.grid()

    if show_values:
        for axis in axes:
            for bar in axis:
                w, h = bar.get_width(), bar.get_height()
                plt.text(bar.get_x() + w/2, bar.get_y() + h/2, 
                         value_format.format(h), ha="center", 
                         va="center")
```

```python
plt.figure(figsize=(8, 4))

course = ['Secured', 'Lost']

data = [
    [60, 70, 80, 90],
    [40, 30, 20, 10]
]

category_labels = ['Calculus', 'History', 'Algorithm', 'Statistics']

plot_stacked_bar(
    data, 
    course, 
    category_labels=category_labels, 
    show_values=True, 
    value_format="{:.1f}",
    colors=['tab:green', 'tab:orange'],
    y_label="percentage"
)

plt.savefig('bar.png')
plt.show()
```

```python
# Creating a DataFrame
# Creating lists for different columns
list_labels = ['Description','Secured', 'Total', 'Weight in %']

Description = ['First Quiz', 'Second Quiz', 'Fassignment', 'Sassignment','Midterm', 'Final']

Secured = [98, 97, 95.5, 96, 98, 0]
Total = [100, 100, 100, 100, 100, 100]
Weight =[10, 10, 10, 10, 30, 30]

list_cols = [Description, Secured,Total, Weight]
zipped = list(zip(list_labels, list_cols))
data = dict(zipped)
score_df = pd.DataFrame(data)
```

```python
# Creating DataFrame for Grades
list_labels = ['Letter Grade','lRange', 'uRange'] # labels/header for each column

Description = ['A','A-','B+','B', 'B-', 'C+', 'C', 'C-', 'D','F'] # first column
lRange = [93, 90, 87, 83,80,77,73,70,60,0 ] # second column
uRange = [100, 92.9, 89.9, 86.9, 82.9, 79.9, 76.9, 72.9, 69.9, 60] # third column

list_cols = [Description, lRange, uRange] #list of columns [1st col, 2nd col, 3rd col], a list of lists
zipped = list(zip(list_labels, list_cols)) # converts the zip to datatype:list
#Python's zip() function is defined as zip(*iterables) . 
#The function takes in iterables as arguments and returns an iterator. 
#This iterator generates a series of tuples containing elements from each iterable. 
# zip() can accept any type of iterable, such as files, lists, tuples, dictionaries, sets, and so on
data2 = dict(zipped)

grade_df= pd.DataFrame(data2)

type(zip(list_labels, list_cols))
```

```python
temp_df = score_df.copy()
temp_df["Secured %"] = (score_df["Secured"]/score_df["Total"]*score_df["Weight in %"]).round(2)
temp_df.rename(columns ={'Secured':'Raw Score'}, inplace  = True)
temp_df
```

```python
total_score =temp_df["Secured %"].sum()
remaining_score = temp_df.loc[temp_df['Raw Score'] == 0]["Weight in %"].sum()
max_score = total_score+remaining_score
letter_grade =grade_df.loc[(grade_df['lRange'] <= max_score) & (grade_df['uRange'] >= max_score)]['Letter Grade'].values[0]


print ("Your current score is: ", total_score, '%')
print ("The maximum you can score is", remaining_score, "%", "for a total of ", max_score , "%")
print ("which will give u a letter grade of ", letter_grade) # code to extract letter grade later
```

```python
Secured_t = temp_df.loc[temp_df['Raw Score'] != 0]["Secured %"].sum()
Attempted_t = temp_df.loc[temp_df['Raw Score'] != 0]["Weight in %"].sum()
lost_points = (Attempted_t - Secured_t).round(2)
```

```python
blue = 93-(total_score)
grey = 100 - lost_points-93

Secured_t = temp_df.loc[temp_df['Raw Score'] != 0]["Secured %"].sum()
Attempted_t = temp_df.loc[temp_df['Raw Score'] != 0]["Weight in %"].sum()
lost_points = (Attempted_t - Secured_t).round(2)


plt.figure(figsize=(8,8))
labels = ['Earned Points',  'Lost Points', 'Possible Points','Points needed']


sizes = [total_score, lost_points,grey, blue]
colors = ['yellowgreen','red', 'silver', 'lightskyblue']
explode = (0, 0, 0, 0)  

# Plot
plt.pie(sizes, labels=labels, colors=colors,labeldistance = 1.4, pctdistance=1.2,
autopct='%1.2f%%', shadow=True, startangle=230)
plt.title ("To Earn a Grade of A", pad = 70)

plt.legend(bbox_to_anchor=(1.3, 0.3), loc='upper left')
plt.show()
```
